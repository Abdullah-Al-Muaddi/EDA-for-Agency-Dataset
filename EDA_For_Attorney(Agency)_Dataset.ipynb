###**Convert** data types

**Convert** data type from object ---> int

```python
df_attorney["عدد الوكالات الصادرة"] = df_attorney["عدد الوكالات الصادرة"].str.replace(',', '')
df_attorney["عدد الوكالات الصادرة"] = df_attorney["عدد الوكالات الصادرة"].astype(int)
```

**Convert** data type from object ---> date

```python
df_attorney["التاريخ ميلادي"]=pd.to_datetime(df_attorney["التاريخ ميلادي"])
```

### **Add** new columns:

**Add month(الشهر) , day(اليوم) and weekdays(ايام الاسبوع)**

```python
df_attorney["الشهر"] = df_attorney["التاريخ ميلادي"].dt.month
df_attorney["اليوم"] = df_attorney["التاريخ ميلادي"].dt.day
df_attorney["ايام الاسبوع"] = df_attorney["التاريخ ميلادي"].dt.day_name()
df_attorney.head()
```



---

### **this function used to fix the arabic text written in the plots**

```python
import arabic_reshaper
revase_dict ={}#to increase runtime
# --->it saves the prevous words in a dict so we don't have to call the reshaper every time
def reshape_revase_arabic(arabic_string_list):
    if(arabic_string_list not in revase_dict ):
       reshape =arabic_reshaper.reshape(arabic_string_list)
       reverse = reshape[::-1] #to make it from right to left
       revase_dict[arabic_string_list] = reverse
    else:
       reverse = revase_dict[arabic_string_list]
    return reverse
```



---



### **uses the reshape_revase_arabic method for every arabic column**


**To ensure that arabic text viewed in a good form in plots**

```python
convert_features = ["المنطقة","المدينة","نوع القطاع","نوع الخدمة","نوع الوكالة"]
df_attorney_new = df_attorney.copy()
for feature in convert_features:
  df_attorney_new[feature] = df_attorney_new[feature].apply(reshape_revase_arabic)
```

**plot showing the distribution of Agencies issued from each day of the month**
